```python
result_query = collection.find({})
list_query = []
for item in result_query:
    list_query.append(item)
df_query = pd.DataFrame(list_query)
```

```python
def gen_consistent(df_query,list_consist):
    dict_rela = {}

    for item in list_consist:
        slave = item[0]
        boss = item[1]
        
        key_rela = slave + '_' + boss
        dict_career_consist = {}
        for i in range(len(df_query)):
            major_code = df_query[boss].iloc[i][0]
            career = df_query[slave].iloc[i]
            dict_career_consist[major_code] = career
        dict_rela[key_rela] = dict_career_consist
    return dict_rela
```

```python
list_consist = [['tuition','type_edu'], ['career','major_code'],['object','case'],['register','case']]
dict_consist = gen_consistent(df_query,list_consist)
```
